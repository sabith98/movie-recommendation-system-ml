```python
movies = movies[['movie_id','original_title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast

def extract_names_to_list(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L 
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(extract_names_to_list)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(extract_names_to_list)
movies.head()
```

```python
def extract_three_names_to_list(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(extract_three_names_to_list)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def extract_director_names_to_list(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(extract_director_names_to_list)
```

```python
movies.sample(5)
```

```python
def remove_spaces(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(remove_spaces)
movies['crew'] = movies['crew'].apply(remove_spaces)
movies['genres'] = movies['genres'].apply(remove_spaces)
movies['keywords'] = movies['keywords'].apply(remove_spaces)
```

```python

movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['original_title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['original_title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].original_title)
```

```python
recommend('Avatar')
```

### Using TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer(max_features=5000,stop_words='english')

# vector = cv.fit_transform(new['tags']).toarray()
tfidf_matrix = vectorizer.fit_transform(new['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
# cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
def recommend_movies(user_profile, cosine_sim=cosine_sim):
    indices = pd.Series(new.index, index=new['original_title']).drop_duplicates()
    sim_scores = list(enumerate(cosine_sim[indices[user_profile]]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:3]
    movie_indices = [i[0] for i in sim_scores]
    return new['original_title'].iloc[movie_indices]

# Example usage
print(recommend_movies("avatar"))
```
